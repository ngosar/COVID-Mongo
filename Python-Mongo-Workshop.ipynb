```python
import pymongo #import general library
from pymongo import MongoClient 
```

```python
url= 'mongodb+srv://readonly:readonly@covid-19.hip2i.mongodb.net/covid19'
client = MongoClient(url)
```

```python
client.list_database_names()
```

```python
covid19_db = client.get_database('covid19')
```

```python
covid19_db.list_collection_names()
```

```python
global_and_us = covid19_db['global_and_us']
```

```python
global_and_us.find_one()
```

```python
#covid19_db['metadata'].find_one()
```

```python
for r in global_and_us.find().limit(5):#find will retrieve everything without the limit argument
    print(r)
```

```python
for r in global_and_us.find().limit(5):#find will retrieve everything without the limit argument
    print(r['country'])#r is all dictionarys, if only want country can query key

for r in global_and_us.find({},{'_id':0, 'country':1}).limit(5): #find.({filters, must put this first pair of brackets},{projections})
    print(r)
```

```python
for r in global_and_us.find({'country':'Ireland'}):
    print(r)
```

```python
import datetime

for r in global_and_us.find({'date': datetime.datetime(2020, 1,23,0,0)}):
    print(r)
```

```python
global_and_us.distinct('country')
```

```python
for agg in global_and_us.aggregate([
        {'$group':{'_id':'$country','count':{'$sum': 1}}},
        {'$sort' : { 'count' : -1, '_id': 1 } },
    ]):
    print(agg)
```

```python
for agg in global_and_us.aggregate([
        {'$group':{'_id':'$country','recovered':{'$sum': '$recovered'}}},
    ]):
    print(agg)
```
